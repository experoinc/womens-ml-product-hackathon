```python
import numpy as np
import pandas as pd
```

# Read in the data and have a first look

Make sure your character encodings are set to `latin1`, because these files are a bit wonky on the binary encoding for `utf-8`.

```python
andesite    = pd.read_csv('csv/ANDESITE_part1.csv', encoding='latin1')
basalt      = pd.read_csv('csv/BASALT_part1.csv', encoding='latin1')
carbonatite = pd.read_csv('csv/CARBONATITE.csv', encoding='latin1')
kimberlite  = pd.read_csv('csv/KIMBERLITE.csv', encoding='latin1')
rhyolite    = pd.read_csv('csv/RHYOLITE_part1.csv', encoding='latin1')
```

```python
print(andesite.shape)
print(basalt.shape)
print(carbonatite.shape)
print(kimberlite.shape)
print(rhyolite.shape)
```

```python
with pd.option_context('display.max_columns', None):
    display(andesite.head())
```

# Clean the dataframes for numpy array prep

If you look at the files using .tail(), you'll find that the end members of the files are references to papers. We want to drop this info, since we want exclusively rock samples in our data.

```python
carbonatite.tail()
```

Every file has a line populated only by the text, "References:", so we'll search for that line, and drop all indices including and after that.

```python
reference_index = andesite[andesite['CITATIONS'] == 'References:'].index[0]
andesite.drop(range(reference_index, andesite.shape[0]), axis=0, inplace=True)

reference_index = basalt[basalt['CITATIONS'] == 'References:'].index[0]
basalt.drop(range(reference_index, basalt.shape[0]), axis=0, inplace=True)

reference_index = carbonatite[carbonatite['CITATIONS'] == 'References:'].index[0]
carbonatite.drop(range(reference_index, carbonatite.shape[0]), axis=0, inplace=True)

reference_index = kimberlite[kimberlite['CITATIONS'] == 'References:'].index[0]
kimberlite.drop(range(reference_index, kimberlite.shape[0]), axis=0, inplace=True)

reference_index = rhyolite[rhyolite['CITATIONS'] == 'References:'].index[0]
rhyolite.drop(range(reference_index, rhyolite.shape[0]), axis=0, inplace=True)
```

Let's now take a look at the new lengths of the dataframes and compare them to the old shapes to see how many rows we've dropped:

```python
print(andesite.shape)
print(basalt.shape)
print(carbonatite.shape)
print(kimberlite.shape)
print(rhyolite.shape)
```

Now it's time to drop some unneeded columns. Let's build a list of columns we don't want to use in our deep learning algorithm, so that we can drop this information from our dataframes:

```python
columns_to_drop = list(andesite.columns[:27])
columns_to_drop.extend(andesite.columns[-2:])
print(columns_to_drop)
```

Next, let's actually drop this info:

```python
andesite.drop(columns_to_drop, axis=1, inplace=True)
basalt.drop(columns_to_drop, axis=1, inplace=True)
carbonatite.drop(columns_to_drop, axis=1, inplace=True)
kimberlite.drop(columns_to_drop, axis=1, inplace=True)
rhyolite.drop(columns_to_drop, axis=1, inplace=True)
```

We can see from the above `.head()` call that we have a ton of `NaN`s in our data. This will tend to destroy our deep learning algo, because propagating `NaN`s through a neural network results in `NaN` output.

Remember, neural networks are simply matrix multiplication and addition. If we pass a `NaN` through a linear system of equations, we get a `NaN` out. This problem is magnified in the backward direction (during gradient backpropagation) because the `NaN` information is spread out into the system of equations through the gradients.

```python
andesite.fillna(value=0, inplace=True)
basalt.fillna(value=0, inplace=True)
carbonatite.fillna(value=0, inplace=True)
kimberlite.fillna(value=0, inplace=True)
rhyolite.fillna(value=0, inplace=True)
```

```python
with pd.option_context('display.max_columns', None):
    display(kimberlite.head())
```

# Build labels

Now that we have a clean set of data (feature vectors), we'll build a set of corresponding labels we can train against. We have five rock types, so we'll use five indices (one for each type):

* andesite: 0
* basalt: 1
* carbonatite: 2
* kimberlite: 3
* rhyolite: 4

```python
label = pd.Series([0 for i in range(andesite.shape[0])])
andesite['label'] = label

label = pd.Series([1 for i in range(basalt.shape[0])])
basalt['label'] = label

label = pd.Series([2 for i in range(carbonatite.shape[0])])
carbonatite['label'] = label

label = pd.Series([3 for i in range(kimberlite.shape[0])])
kimberlite['label'] = label

label = pd.Series([4 for i in range(rhyolite.shape[0])])
rhyolite['label'] = label
```

# Build one big ass matrix

Now let's concatenate all this information into one array which we'll write out and use for training and testing in the next section.

```python
rocks = andesite
rocks = rocks.append(basalt, ignore_index=True)
rocks = rocks.append(carbonatite, ignore_index=True)
rocks = rocks.append(kimberlite, ignore_index=True)
rocks = rocks.append(rhyolite, ignore_index=True)
```

```python
rocks.tail()
```

# Write out our clean data

```python
rocks.to_csv('csv/clean_rocks.csv', index=False)
```
